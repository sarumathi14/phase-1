```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
url = "https://api.worldbank.org/v2/en/indicator/SP.DYN.LE00.IN?format=json&per_page=20000"
years = list(range(1960, 2021, 5))
life_exp = [52.5, 56.1, 59.3, 62.3, 64.5, 66.5, 68.1, 69.5, 70.8, 72.0, 72.6, 72.9, 73.3]

plt.figure(figsize=(12, 6))
sns.lineplot(x=years, y=life_exp, marker='o', linewidth=2.5)

plt.title('Global Average Life Expectancy (1960-2020)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(years, rotation=45)
plt.tight_layout()
plt.show()
```
