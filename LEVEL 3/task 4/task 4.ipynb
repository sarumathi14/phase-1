```python
import numpy as np
from scipy import stats
group1 = np.array([5, 7, 8, 6, 9, 7, 6, 8, 7, 6])
group2 = np.array([6, 8, 9, 7, 10, 8, 7, 9, 8, 7])
print("Group 1 Mean:", np.mean(group1))
print("Group 2 Mean:", np.mean(group2))
print("Group 1 Std Dev:", np.std(group1, ddof=1))
print("Group 2 Std Dev:", np.std(group2, ddof=1))
t_stat, p_value = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the groups.")
```
