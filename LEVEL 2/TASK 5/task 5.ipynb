```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic = pd.read_csv(r'C:\Users\sarum\OneDrive\Desktop\LEVEL 2\TASK 5\Titanic-Dataset.csv') 
print("Numeric columns in dataset:")
print(titanic.select_dtypes(include=['number']).columns.tolist())
correlation_matrix = titanic.select_dtypes(include=['number']).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,  
    cmap='coolwarm', 
    center=0,  
    fmt='.2f', 
    linewidths=0.5, 
    annot_kws={'size': 10} )
plt.title('Titanic Dataset Correlation Heatmap', fontsize=14, pad=20)
plt.xticks(rotation=45, ha='right')  
plt.tight_layout() 
plt.show()
```
