```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
data = pd.read_csv(r'C:\Users\sarum\OneDrive\Desktop\task 2.txt')
last_5_days = data.tail(5)
model_rmse = np.sqrt(mean_squared_error(last_5_days['actual_temperature'], last_5_days['model_forecast']))
weather_com_rmse = np.sqrt(mean_squared_error(last_5_days['actual_temperature'], last_5_days['weather_com_forecast']))
imd_rmse = np.sqrt(mean_squared_error(last_5_days['actual_temperature'], last_5_days['imd_forecast']))
print(f"Model RMSE: {model_rmse}")
print(f"Weather.com RMSE: {weather_com_rmse}")
print(f"IMD RMSE: {imd_rmse}")
if model_rmse < weather_com_rmse and model_rmse < imd_rmse:
    print("Your model performed the best!")
elif weather_com_rmse < model_rmse and weather_com_rmse < imd_rmse:
    print("Weather.com performed the best!")
else:
    print("IMD performed the best!")
```
