```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
n = 10
p = 0.5  
k_values = np.arange(0, n+1)
pmf_values = binom.pmf(k_values, n, p)
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.title(f'Binomial PMF (n={n}, p={p})')
plt.xlabel('k')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
from scipy.stats import norm
mu = 0  
sigma = 1  
x_values = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x_values, mu, sigma)
plt.plot(x_values, pdf_values, color='red', linewidth=2)
plt.title(f'Normal PDF (μ={mu}, σ={sigma})')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
cdf_values = binom.cdf(k_values, n, p)
plt.step(k_values, cdf_values, where='post', color='green', linewidth=2)
plt.title(f'Binomial CDF (n={n}, p={p})')
plt.xlabel('k')
plt.ylabel('Cumulative Probability')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
