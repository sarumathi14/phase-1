```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic_clean = titanic.drop(['alive', 'who', 'adult_male', 'deck', 'embark_town'], axis=1).dropna()
print("Dataset Shape:", titanic_clean.shape)
print("\nFirst 5 rows:")
print(titanic_clean.head())
print("\nData Types:")
print(titanic_clean.dtypes)
print("\nMissing Values:")
print(titanic_clean.isnull().sum())
print("\nMeasures of Central Tendency:")
print(titanic_clean[['age', 'fare']].describe())
print("\nMode for Categorical Features:")
print(titanic_clean[['sex', 'class', 'embarked', 'alone']].mode())
categorical_features = ['sex', 'class', 'embarked', 'alone', 'pclass', 'sibsp', 'parch']
plt.figure(figsize=(15, 12))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 3, i)
    sns.barplot(x=feature, y='survived', data=titanic_clean)
    plt.title(f'Survival Rate by {feature}')
    plt.ylabel('Survival Rate')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
