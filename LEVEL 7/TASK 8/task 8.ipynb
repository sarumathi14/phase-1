```python
import numpy as np
import matplotlib.pyplot as plt
n_trials = 10        
p_success = 0.5     
n_experiments = 1000 
data = np.random.binomial(n=n_trials, p=p_success, size=n_experiments)
mean = np.mean(data)
theoretical_mean = n_trials * p_success
plt.figure(figsize=(10, 6))
plt.hist(data, bins=range(n_trials+2), density=True, 
         alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, 
            label=f'Simulated Mean: {mean:.2f}')
plt.axvline(theoretical_mean, color='green', linestyle='dashed', linewidth=2, 
            label=f'Theoretical Mean: {theoretical_mean}')
plt.title(f'Binomial Distribution Simulation\n(n={n_trials}, p={p_success}, {n_experiments} experiments)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.xticks(range(n_trials+1))
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
print("Binomial Distribution Simulation Results")
print("---------------------------------------")
print(f"Theoretical mean (n × p): {theoretical_mean}")
print(f"Simulated mean: {mean:.4f}")
print(f"Absolute difference: {abs(mean - theoretical_mean):.4f}")

if abs(mean - theoretical_mean) < 0.1:
    print("\nThe simulation matches the theoretical binomial expectation.")
else:
    print("\nThe simulation differs somewhat from the theoretical expectation.")
```
