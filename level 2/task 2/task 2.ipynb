```python
import pandas as pd
import numpy as np
# Create a sample dataset
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, 3, 4, 5],
    'D': [np.nan, np.nan, np.nan, np.nan, np.nan]
}

# Convert to DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
# Detect missing values
missing_values = df.isnull()
print("\nMissing Values (True indicates missing):")
print(missing_values)
# Count missing values in each column
missing_count = df.isnull().sum()
print("\nNumber of Missing Values in Each Column:")
print(missing_count)
```

```python
# Drop rows with any missing values
df_dropped = df.dropna()
print("\nDataFrame After Dropping Rows with Missing Values:")
print(df_dropped)

# Drop columns with all missing values
df_dropped_columns = df.dropna(axis=1, how='all')
print("\nDataFrame After Dropping Columns with All Missing Values:")
print(df_dropped_columns)
# Fill missing values with a specific value (e.g., 0)
df_filled = df.fillna(0)
print("\nDataFrame After Filling Missing Values with 0:")
print(df_filled)

# Fill missing values with the mean of the column
df_filled_mean = df.fillna(df.mean())
print("\nDataFrame After Filling Missing Values with Column Mean:")
print(df_filled_mean)
```
