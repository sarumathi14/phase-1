```python
from datetime import datetime, timedelta
from airflow import DAG
from airflow.operators.python import PythonOperator
from airflow.models import Variable

default_args = {
    'owner': 'airflow',
    'depends_on_past': False,
    'start_date': datetime(2023, 1, 1),
    'email_on_failure': False,
    'email_on_retry': False,
    'retries': 1,
    'retry_delay': timedelta(minutes=5),
}

def print_api_endpoint():
    # Get the API endpoint from AirFlow Variables
    endpoint = Variable.get("api_endpoint")
    print(f"API Endpoint: {endpoint}")

with DAG(
    'api_endpoint_dag',
    default_args=default_args,
    description='A simple DAG that demonstrates using AirFlow Variables',
    schedule_interval=timedelta(days=1),
    catchup=False,
) as dag:
    
    print_endpoint_task = PythonOperator(
        task_id='print_api_endpoint',
        python_callable=print_api_endpoint,
    )

    print_endpoint_task
```
