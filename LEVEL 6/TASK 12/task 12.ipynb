```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
np.random.seed(42)
normal_data = np.random.normal(loc=0, scale=1, size=100)
skewed_data = np.random.exponential(scale=1, size=100)
mixed_data = np.concatenate([normal_data, skewed_data])
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
stats.probplot(normal_data, dist="norm", plot=plt)
plt.title("QQ Plot - Normal Data")
plt.grid(True)
plt.subplot(1, 3, 2)
stats.probplot(skewed_data, dist="norm", plot=plt)
plt.title("QQ Plot - Skewed Data")
plt.grid(True)
plt.subplot(1, 3, 3)
stats.probplot(mixed_data, dist="norm", plot=plt)
plt.title("QQ Plot - Mixed Data")
plt.grid(True)
plt.tight_layout()
plt.show()
```
